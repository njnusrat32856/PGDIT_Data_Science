```python
import pandas as pd

df = pd.read_csv("sales_record - sales_record.csv")

print(df.head())
```

```python
print("=== General Info ===")
df.info()

print("\n=== Summary Statistics ===")
print(df.describe())
```

```python
print("=== Unique Regions ===")
print(df['Region'].unique())

print("\n=== Unique Countries ===")
print(df['Country'].unique())

print("\n=== Unique Item Types ===")
print(df['Item Type'].unique())

print("\n=== Unique Sales Channels ===")
print(df['Sales Channel'].unique())

print("\n=== Unique Order Priorities ===")
print(df['Order Priority'].unique())
```

```python
total_records = len(df)
print("Total number of sales records:", total_records)
```

```python
filtered_df = df[(df['Region'] == 'Sub-Saharan Africa') & (df['Sales Channel'] == 'Online')]

print(filtered_df)
```

```python
result = filtered_df[['Country', 'Item Type', 'Units Sold']]

print(result)
```

```python
high_priority_orders = df[(df['Order Priority'] == 'H') & (df['Units Sold'] > 5000)]

print(high_priority_orders)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

print(df.dtypes[['Order Date', 'Ship Date']])
```

```python
df['Shipping Delay (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days

delayed_orders = df[df['Shipping Delay (Days)'] > 10]

print(delayed_orders)
```

```python
df['Total Revenue'] = df['Units Sold'] * df['Unit Price']

print(df[['Units Sold', 'Unit Price', 'Total Revenue']].head())
```

```python
df['Total Cost'] = df['Units Sold'] * df['Unit Cost']

print(df[['Units Sold', 'Unit Cost', 'Total Cost']].head())
```

```python
df['Profit'] = df['Total Revenue'] - df['Total Cost']

print(df[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Profit']].head())
```

```python
df['Profit Margin (%)'] = round((df['Profit'] / df['Total Revenue']) * 100, 2)

print(df[['Total Revenue', 'Total Cost', 'Profit', 'Profit Margin (%)']].head())
```

```python
region_revenue = df.groupby('Region')['Total Revenue'].sum().reset_index()

print(region_revenue)
```

```python
avg_profit_item = df.groupby('Item Type')['Profit'].mean().reset_index()

print(avg_profit_item)
```

```python
item_sales = df.groupby('Item Type')['Units Sold'].sum().reset_index()

item_sales = item_sales.sort_values(by='Units Sold', ascending=False)

print(item_sales)

top_item = item_sales.iloc[0]
print("\nMost Sold Item Type:", top_item['Item Type'])
print("Total Units Sold:", top_item['Units Sold'])
```

```python
country_profit = df.groupby('Country')['Profit'].sum().reset_index()

country_profit = country_profit.sort_values(by='Profit', ascending=False)

top3_countries = country_profit.head(3)
print(top3_countries)
```
